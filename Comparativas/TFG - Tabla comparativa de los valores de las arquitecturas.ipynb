```python
import csv
import pandas as pd
from prettytable import PrettyTable
```

```python
metrics_training_ResnetUnet = {}
with open('G:/dataset_urbano/saved_values/test_dataset_urbano_ResNetUNet.csv', 'r') as archive:
    reader = csv.reader(archive)
    for row in reader:
        if len(row) == 2:
            key = row[0]
            value = row[1]
            metrics_training_ResnetUnet[key] = value
```

```python
metrics_training_Deeplabv3 = {}
with open('G:/dataset_urbano/saved_values/test_dataset_urbano_DeepLabV3.csv', 'r') as archive:
    reader = csv.reader(archive)
    for row in reader:
        if len(row) == 2:
            key = row[0]
            value = row[1]
            metrics_training_Deeplabv3[key] = value    
```

```python
equivalence_colors = {
        #Rural dataset classes
        '(127, 127, 63)':"Hill", 
        '(0, 127, 0)':"Forest",
        '(255, 255, 0)':"Residential",
        '(0, 255, 0)':"Land",
        '(255, 0, 255)':"Church",
        '(255, 255, 255)':"Road",
        '(127, 127, 0)':"Fence",
        '(255, 0, 0)':"Person",
        '(127, 127, 127)':"Car",
        '(0, 255, 255)':"Sky",
        '(255, 127, 0)':"Haystack",
        '(0, 0, 255)':"Water",
    
        #Urban dataset classes
        '(0, 0, 0)':"Clutter",
        '(128, 0, 0)':"Building",
        '(128, 64, 128)':"Road",
        '(192, 0, 192)':"Static_Car",
        '(0, 128, 0)':"Tree",
        '(128, 128, 0)':"Vegetation",
        '(64, 64, 0)':"Human",
        '(64, 0, 128)':"Moving_Car",
}

data = []

tp_global_resnetunet = 0
fp_global_resnetunet = 0
tn_global_resnetunet = 0
fn_global_resnetunet = 0

tp_global_deeplabv3 = 0
fp_global_deeplabv3 = 0
tn_global_deeplabv3 = 0
fn_global_deeplabv3 = 0

table = PrettyTable()
table.field_names = ["Nombre Clase", "Precisión ResNetUNet", "Precisión DeepLabV3", "Recall ResNetUNet", "Recall DeepLabV3", "Accuracy ResNetUNet", "Accuracy DeepLabV3", "F1 ResNetUNet", "F1 DeepLabV3", "IOU ResNetUNet", "IOU DeepLabV3"]

resultTable = PrettyTable()
resultTable.field_names = ["Nombre Clase", "Precisión", "Recall", "Accuracy", "F1", "IOU"]


for (category1, data1), (category2, data2) in zip(metrics_training_ResnetUnet.items(), metrics_training_Deeplabv3.items()):

    data1=eval(data1)
    tp_resnetunet = data1["TP"]
    fp_resnetunet = data1["FP"]
    tn_resnetunet = data1["TN"]
    fn_resnetunet = data1["FN"]
    
    tp_global_resnetunet += data1["TP"]
    fp_global_resnetunet += data1["FP"]
    tn_global_resnetunet += data1["TN"]
    fn_global_resnetunet += data1["FN"]

    try:
        precision_resnetunet = tp_resnetunet / (tp_resnetunet + fp_resnetunet)
    except ZeroDivisionError:
        precision_resnetunet = 0
    
    recall_resnetunet = tp_resnetunet / (tp_resnetunet + fn_resnetunet)
    accuracy_resnetunet = (tp_resnetunet + tn_resnetunet) / (tp_resnetunet + tn_resnetunet + fp_resnetunet + fn_resnetunet)
    
    try:
        f1_resnetunet = 2*((precision_resnetunet*recall_resnetunet)/(precision_resnetunet+recall_resnetunet))
    except ZeroDivisionError:
        f1_resnetunet = 0
    
    IOU_resnetunet = tp_resnetunet / (fp_resnetunet + tp_resnetunet + fn_resnetunet)    
    
    data2=eval(data2)
    tp_deeplabv3 = data2["TP"]
    fp_deeplabv3 = data2["FP"]
    tn_deeplabv3 = data2["TN"]
    fn_deeplabv3 = data2["FN"]
    
    tp_global_deeplabv3 += data2["TP"]
    fp_global_deeplabv3 += data2["FP"]
    tn_global_deeplabv3 += data2["TN"]
    fn_global_deeplabv3 += data2["FN"]

    
    try:
        precision_deeplabv3 = tp_deeplabv3 / (tp_deeplabv3 + fp_deeplabv3)
    except ZeroDivisionError:
        precision_deeplabv3 = 0
        
    recall_deeplabv3 = tp_deeplabv3 / (tp_deeplabv3 + fn_deeplabv3)
    accuracy_deeplabv3 = (tp_deeplabv3 + tn_deeplabv3) / (tp_deeplabv3 + tn_deeplabv3 + fp_deeplabv3 + fn_deeplabv3)
    
    try:
        f1_deeplabv3 = 2*((precision_deeplabv3*recall_deeplabv3)/(precision_deeplabv3+recall_deeplabv3))
    except ZeroDivisionError:
        f1_deeplabv3 = 0
    
    IOU_deeplabv3 = tp_deeplabv3 / (fp_deeplabv3 + tp_deeplabv3 + fn_deeplabv3)    
    
    className = equivalence_colors[category1]
    
    table.add_row([className, f"{precision_resnetunet:.4f}",  f"{precision_deeplabv3:.4f}", f"{recall_resnetunet:.4f}", f"{recall_deeplabv3:.4f}", f"{accuracy_resnetunet:.4f}", f"{accuracy_deeplabv3:.4f}", f"{f1_resnetunet:.4f}", f"{f1_deeplabv3:.4f}", f"{IOU_resnetunet:.4f}", f"{IOU_deeplabv3:.4f}"])


    
precision_global_resnetunet = tp_global_resnetunet / (tp_global_resnetunet + fp_global_resnetunet)
recall_global_resnetunet = tp_global_resnetunet / (tp_global_resnetunet + fn_global_resnetunet)
accuracy_global_resnetunet = (tp_global_resnetunet + tn_global_resnetunet) / (tp_global_resnetunet + tn_global_resnetunet + fp_global_resnetunet + fn_global_resnetunet)
f1_global_resnetunet = 2*((precision_global_resnetunet*recall_global_resnetunet)/(precision_global_resnetunet+recall_global_resnetunet))
iou_global_resnetunet = tp_global_resnetunet / (fp_global_resnetunet + tp_global_resnetunet + fn_global_resnetunet)

precision_global_deeplabv3 = tp_global_deeplabv3 / (tp_global_deeplabv3 + fp_global_deeplabv3)
recall_global_deeplabv3 = tp_global_deeplabv3 / (tp_global_deeplabv3 + fn_global_deeplabv3)
accuracy_global_deeplabv3 = (tp_global_deeplabv3 + tn_global_deeplabv3) / (tp_global_deeplabv3 + tn_global_deeplabv3 + fp_global_deeplabv3 + fn_global_deeplabv3)
f1_global_deeplabv3 = 2*((precision_global_deeplabv3*recall_global_deeplabv3)/(precision_global_deeplabv3+recall_global_deeplabv3))
iou_global_deeplabv3 = tp_global_deeplabv3 /  (fp_global_deeplabv3 + tp_global_deeplabv3 + fn_global_deeplabv3)



table.add_row(["Global", f"{precision_global_resnetunet:.4f}", f"{precision_global_deeplabv3:.4f}", f"{recall_global_resnetunet:.4f}",  f"{recall_global_deeplabv3:.4f}",  f"{accuracy_global_resnetunet:.4f}", f"{accuracy_global_deeplabv3:.4f}", f"{f1_global_resnetunet:.4f}", f"{f1_global_deeplabv3:.4f}", f"{iou_global_resnetunet:.4f}", f"{iou_global_deeplabv3:.4f}"])

table
```
